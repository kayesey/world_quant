```python
#Reading data contents from a file
#Read the contents from a stored file

try:
    strFile = open('./data/sample.txt','r')
    strContents =strFile.read()
    #close the file after its been read
    strFile.close()
    
    #print the contents
    print(strContents)
except FileNotFoundError:
    print('File not found. Program exited')
```

```python
#using the with option to open and close a file in python
try:
    strPath = './data/sample.txt'
    with open(strPath,'r') as fileRead:
        print(fileRead.readlines())
except:
    print('An error occurred. Please correct it')
```

```python
#write new content to a file
try:
    strPath ='./data/sample.txt'
    with open(strPath,'w') as strReader:
        #write content to file
        strContent = 'Have to start OSCP and AWAE'
        strReader.write(strContent)
except:
    print('Error occurred. Please correct it')
```

```python
#list all files and directories 
import os as os_file

try:
    
    strPath ='/home/jovyan/datacourse'
    #walk through the paths and prints the lists
    for strDir,_, strFiles in os_file.walk(strPath):
        for fileNames in strFiles:
            getFiles = os_file.path.join(strDir, fileNames)
            print(getFiles)
except:
    print('File or Directory does not exist')
    
            
```
